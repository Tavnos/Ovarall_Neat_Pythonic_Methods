```python
############################################################################ Specific byte format,
load_a_byte = {'magic':[],'ascii':{},'byte':[]}                            # with length 1 bytes,
olbl = []                                                                  # from 0x00 to 0xFF,  
for i in range(0x10):                                                      # and their result,
    p_hold = "b'{}'".format(hex(i))                                        # when forced to eval, 
    olbl += [p_hold.replace('0x','\\x0')]                                  # giving len()                 
for i in range(0xff):                                                      # index()               
    olbl += ["b'\\{}'".format(hex(i+16)).replace('0x','x')]                # total length of byte 
for i in olbl:                                                             # in magic ASCII
    if len(str(eval(i))) == 5:                                             # len(str(eval(i)))
        load_a_byte['magic'] += [(i, eval(i))]                             # len(eval(i))
for i in olbl:                                                             # Byte index
    if len(str(eval(i))) == 4:                                             # and i
        load_a_byte['ascii'][bytes.decode(eval(i), 'ascii')] = i, eval(i)  #
for i in olbl:                                                             #
    if len(str(eval(i))) == 7:                                             #
        load_a_byte['byte'] += [(i, eval(i))]                              #
############################################################################
#      (158+94+4)-1', int(0xFF) == (158+94+4)-1)                           #                                                                #     
############################################################################
print(len(load_a_byte['magic'])) # 4                                     #
print(len(load_a_byte['ascii'])) # 94                                    #
print(len(load_a_byte['byte'])) #  158                                   #
print('a loaded byte from 0x00 to 0xFF sums up our length result: ')        #
```
