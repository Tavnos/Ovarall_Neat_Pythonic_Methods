```python
function = 'x'
def calling_func(x, function=function):
    return eval(function)
```

```python
def sketch_graph(function=function, tangent=None, xlim=None, ylim=None):
    df = pd.DataFrame()
    df['x'] = np.arange(xlim[0],xlim[1],(xlim[1]-xlim[0])/100)
    df['y'] = eval(function.replace('x', "df['x']"))
    df = df.set_index('x')
    df.y.plot(ylim=ylim)
    plt.axhline(0, color='black')
    plt.axvline(0, color='black')
```

```python
def find_limit(x, function=function):
    lower_limit = calling_func(x - .00000001, function=function)
    upper_limit = calling_func(x + .00000001, function=function)
    if np.abs(upper_limit - lower_limit) > 100000:
        print('No limit exists')
        return None
    else:
        L = (lower_limit + upper_limit)/2
        return round(L, 4)
```

```python
def differentiate_derivative(x, function=function):
    h_run = 1/1e10  
    rise = calling_func(x+h_run, function=function) - calling_func(x, function=function)
    return round(rise/h_run,4)
#diff_x = differentiate_derivative(-1, 'x**2')
#print("slope of tangent at x = -1 : {}".format(diff_x))
```

```python
x = sp.symbols('x', real=True, use_latex=False)
y = 6*x*y**2 + x*sp.sin(z) + x**3 + z**8*y
```

```python
def summarize_integral(xstart, xfinish, function=function, rectangles=100000):
    width = (xfinish - xstart)/rectangles
    running_sum = 0
    for i in range(int(rectangles)):
        try:
            height = calling_func(xstart + i*width, function=function)
            # each rectangle is 1 width after the previous one
            area = height * width
            running_sum += area
        except:
            continue
    return round(running_sum,4)
print("sum of area = {}".format(summarize_integral(-1,1, 'x**2')))
```
