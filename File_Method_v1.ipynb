```python
class Manage_Documents:
    def make_file(self, f_name, f_content=''):
        save_txt = open('{}'.format(f_name), 'w')
        save_txt.write(f_content)
        save_txt.close()
    def open_file(self, f_name):
        save_txt = open('{}'.format(f_name), 'r')
        str_read = save_txt.read()
        save_txt.close()
        return str_read
    def append_file(self, f_name, append_str):
        str_read = self.open_file(f_name)
        str_read = str_read + '\n' + append_str
        save_txt = open('{}'.format(f_name), 'w')
        save_txt.write(str_read)
        save_txt.close()
    def ls_dir_file(self, directory=None, filetype='any'):
        ls_dir_raw = os.listdir(directory)
        ls_dir = []
        for i in ls_dir_raw:
            if filetype == 'any':
                ls_dir += [i]
            elif filetype in i:
                ls_dir += [i]
        return ls_dir
    def rename_file(self, name_contains, name_renamed, filetype='.txt', directory=None):
        list_files = self.ls_dir_file(directory, filetype)
        for i in list_files:
            if name_contains in i:
                old_name = i
                new_name = i.replace(i, i.replace(name_contains, name_renamed))
                os.rename(old_name, new_name)
call_md = Manage_Documents()
```
